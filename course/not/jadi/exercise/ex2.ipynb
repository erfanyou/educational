```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr

l=[]
for i in df.columns:
    correlation, p_value = pearsonr(df[i], df['output'])

    print(F"ضریب همبستگی پیرسون {i}:", correlation)
    if p_value<= 0.05 :
        print('has')
    elif p_value >0.05:
        print('has not')
        l.append(i)
    print(2*'\n')


for i in df.columns:
    correlation, p_value = spearmanr(df[i], df['output'])

    print(F"ضریب همبستگی اسپیرمن {i}:", correlation)
    if p_value<= 0.05 :
        print('has')
    elif p_value >0.05:
        print('has not')
        l.append(i)
    print(2*'\n')
l=set(l)
```

```python
from sklearn import preprocessing
c=set(df.columns)
cu=c-l
ndf=df[list(cu)]

X = ndf.drop(columns=['output']).values
y = df['output'].values

x = preprocessing.StandardScaler().fit(X).transform(X.astype(float))

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

k_values = range(1, 31)
cross_val_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    cross_val_scores.append(np.mean(scores))

best_k = k_values[np.argmax(cross_val_scores)]
knn= KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, classification_report

yhat = knn.predict(X_test)

accuracy = accuracy_score(y_test, yhat)

f1 = f1_score(y_test, yhat, average='weighted')

conf_matrix = confusion_matrix(y_test, yhat)
sns.heatmap(conf_matrix, annot=True, fmt='d')
plt.show()

class_report = classification_report(y_test, yhat)

print(f"Accuracy: {accuracy}",3*'\n')
print(f"F1 Score (weighted): {f1}",3*'\n')
print("Confusion Matrix:",3*'\n')
print("Classification Report:",3*'\n')
print(class_report,3*'\n')
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, knn.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```

```python
Ks = k_values[-1]
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

```python
plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.fill_between(range(1,Ks),mean_acc - 3 * std_acc,mean_acc + 3 * std_acc, alpha=0.10,color="green")
plt.legend(('Accuracy ', '+/- 1xstd','+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
```
